```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Define the range of years to scrape (1970 to 2018)
start_year = 1985
end_year = 2024

# Base URL
base_url = "https://indexes.nikkei.co.jp/en/nkave/statistics/dataload"

# List to hold all data
all_data = []

# Headers for the table (we'll extract these dynamically)
headers = []
```

```python
# Iterate through each year and month
for year in range(start_year, end_year + 1):
    for month in range(1, 13):
        print(f"Scraping data for {year}-{month:02d}...")

        # Build the URL
        url = f"{base_url}?list=daily&year={year}&month={month}"

        # Send a GET request to the URL
        response = requests.get(url)

        # Check if the request was successful
        if response.status_code != 200:
            print(f"Failed to retrieve data for {year}-{month:02d}.")
            continue

        # Parse the HTML content
        soup = BeautifulSoup(response.text, "html.parser")

        # Find the table
        table = soup.find("table")
        if not table:
            print(f"No table found for {year}-{month:02d}.")
            continue

        # Extract headers (only once)
        if not headers:
            headers = [th.text.strip() for th in table.find_all("th")]
            headers = ["Year", "Month"] + headers

        # Extract rows of data
        rows = table.find_all("tr")[1:]  # Skip the header row
        for row in rows:
            values = [td.text.strip() for td in row.find_all("td")]
            if values:
                all_data.append([year, month] + values)

# Convert the collected data to a DataFrame
df = pd.DataFrame(all_data, columns=headers)

# Save to a CSV file
output_filename = "nikkei_225_data.csv"
df.to_csv(output_filename, index=False)
print(f"Data saved to '{output_filename}'.")
```
